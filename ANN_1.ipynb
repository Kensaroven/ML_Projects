```python
from sklearn.model_selection import GridSearchCV
from keras.wrappers.scikit_learn import KerasClassifier
```

```python
def create_model(neurons=1):
  model = Sequential()
  model.add(Dense(neurons, input_dim = 8, kernel_initializer='uniform', activation = 'sigmoid'))
  model.add(Dense(1, kernel_initializer='uniform', activation = 'sigmoid'))
  model.compile(loss = 'binary_crossentropy', optimizer = 'rmsprop', metrics = ['accuracy'])
  return model
```

```python
model  = KerasClassifier(build_fn = create_model, epochs = 100, batch_size = 10, verbose = 2)
```

```python
neurons = [1,5,10,15,20,25,30]
param_grid = dict(neurons = neurons)
```

```python
grid = GridSearchCV(estimator = model, param_grid = param_grid)
```

```python
grid_result = grid.fit(X,y)
```

```python
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
```

```python
grid_result.cv_results_
```

```python
mean = grid_result.cv_results_['mean_test_score']
stds = grid_result.cv_results_['std_test_score']
params = grid_result.cv_results_['params']

for mean, stdev, param in zip(mean, stds, params):
  print("%f (%f) with %r" % (mean, stdev, param))
```
